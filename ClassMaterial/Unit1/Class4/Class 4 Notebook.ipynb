```python
df['PartofWeek'] = np.where(df['day_of_week'].isin(['Saturday', 'Sunday']), 'Weekend', 'Weekday')
```

```python
conditions = [
    df['visitors'] < 20,
    df['visitors'].between(21, 30),
    df['visitors'] > 30
]

results = [
    'Low Traffic',
    'Medium Traffic',
    'High Traffic'
]

df['Traffic'] = np.select(conditions, results, 'Other')
```
