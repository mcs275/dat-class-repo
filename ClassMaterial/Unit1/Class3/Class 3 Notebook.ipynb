```python
conditions = [
    (df['quarter'].isin([2, 3])) & (df['PartOfWeek'] == 'Weekend'),
    (df['quarter'].isin([2, 3])) & (df['PartOfWeek'] == 'Weekday'),
    (df['quarter'].isin([1, 4])) & (df['PartOfWeek'] == 'Weekend'),
    (df['quarter'].isin([1, 4])) & (df['PartOfWeek'] == 'Weekday'),
]

results    = [
    'Weekend Warm',
    'Weekday Warm',
    'Weekend Cold',
    'Weekday Cold'
]

df['WeatherPart'] = np.select(conditions, results, 'Other')
```
