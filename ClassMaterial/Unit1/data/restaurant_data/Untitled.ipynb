```python
import pandas as pd
```

```python
reservations = pd.read_csv('air_reserve.csv')
```

```python
stores = pd.read_csv('air_store_info.csv')
```

```python
visits = pd.read_csv('air_visit_data.csv')
```

```python
dates = pd.read_csv('date_info.csv')
dates['calendar_date'] = pd.to_datetime(dates['calendar_date'])
```

```python
reservations['visit_datetime'] = pd.to_datetime(reservations['visit_datetime']).dt.date
reservations['visit_datetime'] = pd.to_datetime(reservations['visit_datetime'])
```

```python
visits['visit_date'] = pd.to_datetime(visits['visit_date'])
```

```python
reservations = reservations.merge(dates, left_on = 'visit_datetime', right_on = 'calendar_date', how = 'left')
```

```python
reservations = reservations.groupby(['air_store_id', 'visit_datetime'])[['reserve_visitors']].sum().reset_index()
```

```python
visits = visits.merge(reservations, left_on = ['air_store_id', 'visit_date'], right_on = ['air_store_id', 'visit_datetime'], how = 'left')
```
