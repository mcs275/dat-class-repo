```python
movies['UNRATED']=np.where(movies.content_rating.isin(['NOT RATED','APPROVED','PASSED','GP']),True,False)
```

```python
movies['UNRATED']=np.where(movies.content_rating.isin(['NOT RATED','APPROVED','PASSED','GP']),"UNRATED","RATED")
```

```python
movies['NC-17']=np.where(movies.content_rating.isin(['X','TV-MA']),True,False)
```

```python
conditions =  [
    (movies['content_rating'].isin(['NOT RATED','APPROVED','PASSED','GP'])),
     (movies['content_rating'].isin(['X','TV-MA']))
]
                       
results    = [
    'UNRATED',
    'NC-17'
]

movies['New Content Ratings'] = np.select(conditions, results, 'Other')
```
