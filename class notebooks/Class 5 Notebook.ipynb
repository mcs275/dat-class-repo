```python
## certain functions built into agg
df.groupby(['day_of_week','holiday']).agg([
    'count','mean','sum','std'
])
```

```python
np.mean(df['visitors'])
```

```python
df.groupby(['day_of_week','holiday']).agg([
    'count',np.mean,'sum','std', pd.Series.mode
])
```

```python
pd.Series.mode(df['visitors'])
```

```python
def is_missing(series):
    return pd.isnull(series).sum()
```

```python
is_missing(df['visitors'])
```

```python
##can pass a function to the agg command
df.groupby(['day_of_week','holiday']).agg([
    'count',np.mean, 'sum','std',is_missing
])
```

```python
##x in lambda refers to the individual groups in the grouping (i.e. the 7 different groupings for each day)
df.groupby('day_of_week').apply(lambda x: x.mean() )
```

```python
##not every pandas method works on a grouping
```
